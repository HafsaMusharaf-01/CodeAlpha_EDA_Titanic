```python

import pandas as pd
import numpy as np

# Dataset load
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

df.head()
```

```python
print(df.shape)
df.info()
```

```python
df.isnull().sum()
```

```python
df['Survived'].value_counts()
```

```python
df.groupby('Sex')['Survived'].mean()
```

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x='Survived')
plt.title('how many people survived')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```

Key Insights from EDA:

1. Out of 891 passengers, only around 340 survived (38%), while the majority (62%) did not survive.

2. Gender had a major impact on survival - female passengers had a much higher survival rate (74%) compared to male passengers (19%). This suggests a "women and children first" evacuation pattern.

3. Passenger class also strongly affected survival - 1st class passengers had the highest survival rate, while 3rd class passengers had the lowest. This may be linked to cabin location and rescue priority.

4. The majority of passengers were between 20-40 years old, with fewer children and elderly passengers on board.

5. Data Issues: The dataset had missing values in the Age column (177 missing), Cabin column (687 missing), and Embarked column (2 missing). These would need to be handled before further statistical modeling.
